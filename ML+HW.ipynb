```python
df = pd.read_csv('raw_news.csv', index_col=0)
texts = df['text'] #прочитали тексты
texts_list = []
for text in texts:
    texts_list.append(text.lower()) #собрали все тексты в список из строк, где каждая строка -- один текст
txt = []
for t in texts_list:
    txt.append(t.split())  #разбили по пробелам, получился список из строк, где каждая строка -- слово 
    #но теперь это отдельный список, не разбитый по текстам, просто все слова в одном списке -- не годится
print(type(texts_list))    
print(type(txt))
print(type(txt[-1]))

    
    
   
```

```python
for t in texts_list:
    txt = t.split() #разбили по пробелам, получился список из строк, где каждая строка -- слово
for t in txt:
    print(t)
```

```python
cleaner_txt = []
for w in txt:
    cleaner_txt.append(re.sub(r'[\W]+', '', w))
if '' in cleaner_txt:
    cleaner_txt.remove('')
print(cleaner_txt)    
```
